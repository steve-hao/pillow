# `ImageFilter`&nbsp;模块

这个&nbsp;`ImageFilter`&nbsp;模块包含一组预定义的筛选器的定义，这些筛选器可与&nbsp;`Image.filter()`&nbsp;方法。

## 示例：过滤图像

```
from PIL import ImageFilter

im1 = im.filter(ImageFilter.BLUR)

im2 = im.filter(ImageFilter.MinFilter(3))
im3 = im.filter(ImageFilter.MinFilter)  # same as MinFilter(3)
```

_class_ `PIL.ImageFilter.Color3DLUT`(_size_,&nbsp;_table_,&nbsp;_channels=3_,&nbsp;_target_mode=None_,&nbsp;_**kwargs_)

三维颜色查找表。
    
使用通道的值作为三维查找表中的坐标转换3通道像素，并插入最近的元素。
    
此方法允许您使用预先计算的抽取表在恒定时间内应用几乎任何颜色转换。
    
5.2.0 新版功能.
    
* 参数
  *  **size**&nbsp;-- 表格的大小。（int，int，int）的一个int或元组。任何尺寸的最小尺寸为2，最大尺寸为65。
  *  **table**&nbsp;-- 平面查找表。一览表&nbsp;`channels * size**3`&nbsp;浮动元素或&nbsp;`size**3`&nbsp;通道大小的浮点数元组。首先改变通道，然后是第一个维度，然后是第二个维度，然后个维度。值0.0对应输出的最小值，1.0最高。
  *  **channels**&nbsp;-- 表中的通道数。可能是3或4。默认值为3。
  *  **target_mode**&nbsp;-- 结果图像的模式。应该有不少于&nbsp;`channels`&nbsp;渠道。默认是&nbsp;`None`&nbsp;也就是说模式不会改变。
    
* _classmethod_`generate`(_size_,&nbsp;_callback_,&nbsp;_channels=3_,&nbsp;_target_mode=None_)
  
  使用提供的回调生成新的LUT。
        
  *  参数
     *  **size**&nbsp;-- 桌子的大小。传递给构造函数。
     *  **callback**&nbsp;-- 三个参数对应三个颜色通道的功能。将被调用&nbsp;`size**3`&nbsp;值在0.0到1.0之间的时间，应返回一个具有&nbsp;`channels`&nbsp;元素。
     *  **channels**&nbsp;-- 应返回回调的通道数。
     *  **target_mode**&nbsp;-- 传递给结果查找表的构造函数。
    
* `transform`(_callback_,&nbsp;_with_normals=False_,&nbsp;_channels=None_,&nbsp;_target_mode=None_)
   
   使用提供的回调转换表值，并返回具有更改值的新LUT。
        
   * 参数
     *  **callback**&nbsp;-- 获取旧查阅表格值并返回一组新值的函数。函数应采用的参数个数为&nbsp;`self.channels`&nbsp;或&nbsp;`3 + self.channels`&nbsp;如果&nbsp;`withs`&nbsp;已设置标志。应返回的元组&nbsp;`self.channels`&nbsp;或&nbsp;`channels`&nbsp;元素（如果已设置）。
     *  **with_normals**&nbsp;-- 如果是真的，&nbsp;`callback`&nbsp;将以颜色多维数据集中的坐标作为前三个参数调用。否则，&nbsp;`callback`&nbsp;将仅使用实际颜色值调用。
     *  **channels**&nbsp;-- 结果查找表中的通道数。
     *  **target_mode**&nbsp;-- 传递给结果查找表的构造函数。



_class_ `PIL.ImageFilter.BoxBlur`(_radius_)

通过将每个像素设置为每个方向延伸半径像素的方形框中像素的平均值来模糊图像。支持任意大小的浮动半径。使用一个优化的实现，该实现相对于任何半径值的图像大小以线性时间运行。
    
* 参数
  *  **radius**&nbsp;-- 一个方向上的框的大小。半径0不模糊，返回相同的图像。半径1在每个方向上取1个像素，即总共9个像素。



```
im.filter(ImageFilter.BoxBlur(4))
```

* 参数

  *  **percent**&nbsp;-- 不清晰度强度，百分比

```
im.filter(ImageFilter.UnsharpMask(4))
```


_class_ `PIL.ImageFilter.Kernel`(_size_,&nbsp;_kernel_,&nbsp;_scale=None_,&nbsp;_offset=0_)

创建卷积内核。当前版本只支持3x3和5x5整数和浮点内核。
    
在当前版本中，内核只能应用于“l”和“rgb”图像。
    
* 参数
  *  **size**&nbsp;-- 内核大小，表示为（宽度、高度）。在当前版本中，必须是（3,3）或（5,5）。
  *  **kernel**&nbsp;-- 包含核权的序列。
  *  **scale**&nbsp;-- 比例因子如果给定，则每个像素的结果除以该值。默认值是内核权重的总和。
  *  **offset**&nbsp;-- 抵消。如果给定，该值将在除以比例因子后添加到结果中。



```
im1.filter(ImageFilter.Kernel((3,3),(1,1,1,0,0,0,2,0,2)))
```

_class_ `PIL.ImageFilter.RankFilter`(_size_,&nbsp;_rank_)

创建等级筛选器。等级过滤器对给定大小的窗口中的所有像素排序，并返回 rank' 钍值。
    
* 参数
  *  **size**&nbsp;-- 内核大小，以像素为单位。
  *  **rank**&nbsp;-- 要选取的像素值。使用0作为最小过滤器，&nbsp;`size * size / 2`&nbsp;对于中值滤波器，&nbsp;`size * size - 1`&nbsp;对于最大滤波器等。



```
im1.filter(ImageFilter.RankFilter(5,8))
```
