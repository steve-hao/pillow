````
from PIL import Image, ImageEnhance

im = Image.open("cl.jpg")

enhance_im1 = ImageEnhance.Color(im)

# enhance_im1 = ImageEnhance.Contrast(im)
# enhance_im1 = ImageEnhance.Brightness(im)
# enhance_im1 = ImageEnhance.Sharpness(im)

im1 = enhance_im1.enhance(0.5)

im2 = enhance_im1.enhance(1.5)

w, h = im.size

img = Image.new("RGB", (w*3, h))

img.paste(im1, (0, 0))
img.paste(im, (w, 0))
img.paste(im2, (w*2, 0))
img
````
